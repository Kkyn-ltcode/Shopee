```python
def create_gender(names) :
    gender = np.zeros(len(names), dtype=int)
    for i, name in enumerate(names):
        name_splited = word_tokenize(normalize(name))
        is_male = 'nam' in name_splited
        is_female = 'nữ' in name_splited or 'nu' in name_splited
        if is_male :
            if is_female :
                gender[i] = 3
            else :
                gender[i] = 1
        else :
            if is_female :
                gender[i] = 2
            else :
                gender[i] = 3

    return gender
```

```python
def create_category(category, names) :
    category_np = np.zeros(len(names), dtype=int)
    for i, name in enumerate(names) :
        for k, cat in enumerate(category):
            l_cat = cat.split('/')
            for j in l_cat :
                if j in name:
                    category_np[i] = k+1
                    break
    return category_np
```

```python
def normalize(sentence) :
    sentence = unicodedata.normalize("NFC", sentence)
    sentence = sentence.lower()
    sentence = re.sub(r"\s+", ' ', sentence)
    return sentence
```

```python
data['gender'] = data['product_name'].apply(lambda x: 2 if ('nam' in x.lower()) and ('nữ' in x.lower()) else 1 if 'nữ' in x.lower() else 0 if 'nam' in x.lower() else 2)

genders = data['gender'].unique().tolist()

for gender in genders:
    data[gender] = data['gender'].apply(lambda x: 1 if gender == x else 0)
data.rename(columns={0: 'Male', 1: 'Female', 2: 'Unisex'}, inplace=True)
# data['gender'] = data['product_name'].apply(lambda x: create_gender(x))

data['variation_list'] = data['tier_variations'].apply(lambda x: ast.literal_eval(x))
variation_list = data['variation_list'].values.tolist()
color_list = ['màu', 'colo']
size_list = ['kích', 'siz']
for variation in variation_list:
    for prop in variation:
        for i in color_list:
            if i in prop['name'].lower():
                prop['name'] = 'color'
                continue
        for i in size_list:
            if i in prop['name'].lower():
                prop['name'] = 'size'
                continue
        if prop['name'] != 'color' and prop['name'] != 'size':
            prop['name'] = 'none'
            
data['tmp1'] = pd.Series(variation_list).apply(lambda x: '_'.join([i['name'] for i in x]))

data['color'] = data['variation_list'].apply(lambda x: sum([len(variation['options']) for variation in x if variation['name'] == 'color'])).replace(to_replace=0, value=1)

data['size'] = data['variation_list'].apply(lambda x: sum([len(variation['options']) for variation in x if variation['name'] == 'size'])).replace(to_replace=0, value=1)

data['brand'] = data['brand'].replace(to_replace=['No brand', 'NoBrand', '0', 'None'], value='No Brand').fillna(value='No Brand')
data['brand'] = data['brand'].apply(lambda x: 0 if x == 'No Brand' else 1)

data[['price', 'price_min', 'price_max']] = (data[['price', 'price_min', 'price_max']] / 10000).astype(int)

category = ['None', 'kimono', 'kaki', 'jean', 'bomber', 'cardigan', 'sơ mi', 'gió/dù/jacket', 'blazer', 'croptop/gile', 'thể thao', 'áo khoác phao/áo phao', 'nỉ/hoodie', 'denim', 'dạ/lông', 'len','da', 'chống nắng', 'thun', 'lửng']

data['category'] = create_category(category[1:], data.loc[:, 'product_name'])
data['category'] = data['category'].apply(lambda x: category[x])
for cat in category:
    data[cat] = data['category'].apply(lambda x: 1 if x == cat else 0)
data.drop(columns=['category'], inplace=True)
```

```python
def convert_ctime_to_two_time(time_start) :
    nowaday = time.time()
    interval = nowaday-time_start
    time_tmp = interval - 2529000
    if time_tmp <= 0 :
        return 0, int(interval)
    interval -= 2529000
    k = interval // 15768000
    odd = interval % 15768000
    time_summer = 0
    if k % 2 == 0 :
        time_summer = (k//2)*15768000 + odd
    else :
        time_summer = (k//2 + 1) * 15768000
    return int(time_summer), int(interval - time_summer + 2529000)
```

```python
data['two_time'] = data['product_ctime'].apply(lambda x: convert_ctime_to_two_time(x))
data['time_summer'] = data['two_time'].apply(lambda x: x[0])
data['time_winter'] = data['two_time'].apply(lambda x: x[1])
data.drop(columns=['two_time'], inplace=True)
```

```python
data.drop(index=data[data['follower_count'] > 400000].index, inplace=True)
```

```python
train_data, test_data = train_test_split(data, test_size=0.1, random_state=42)

cols = ['shop_location']
target_col = 'historical_sold'
alpha = 2
folds = 4

for col in cols:
    target_mean_global = train_data[target_col].mean()
    nrows_cat = train_data.groupby(col)[target_col].count()
    target_means_cats = train_data.groupby(col)[target_col].mean()
    target_means_cats_adj = (target_means_cats * nrows_cat + 
                             target_mean_global * alpha)/(nrows_cat + alpha)
    encoded_col_test = test_data[col].map(target_means_cats_adj)

    kfold = StratifiedKFold(folds, shuffle=True, random_state=1)
    parts = []
    for tr_in, val_ind in kfold.split(train_data.drop(columns=target_col), train_data[target_col].values):
        df_for_estimation, df_estimated = train_data.iloc[tr_in], train_data.iloc[val_ind]
        nrows_cat = df_for_estimation.groupby(col)[target_col].count()
        target_means_cats = df_for_estimation.groupby(col)[target_col].mean()
        target_means_cats_adj = (target_means_cats*nrows_cat + 
                                 target_mean_global*alpha)/(nrows_cat+alpha)
        encoded_col_train_part = df_estimated[col].map(target_means_cats_adj)
        parts.append(encoded_col_train_part)
    encoded_col_train = pd.concat(parts, axis=0)
    encoded_col_train.fillna(target_mean_global, inplace=True)

    train_data = pd.merge(train_data, 
                          encoded_col_train, 
                          left_index=True, 
                          right_index=True).sort_index().reset_index(drop=True).rename(columns={f'{col}_y': f'{col}_encoded'}).drop(columns=[f'{col}_x'])
    test_data = pd.merge(test_data, 
                         encoded_col_test, 
                         left_index=True, 
                         right_index=True).sort_index().reset_index(drop=True).rename(columns={f'{col}_y': f'{col}_encoded'}).drop(columns=[f'{col}_x'])
col_to_drop = ['product_ctime', 'tier_variations', 'shop_ctime', 'country', 'catid',
               'variation_list', 'tmp1', 'gender', 'itemid', 'shopid', 'item_count',
               'product_name', 'sold', 'shop_name', 'price_min', 'price_max']
data = pd.concat([train_data, test_data], axis=0, ignore_index=True)
data.drop(columns=col_to_drop, inplace=True)
data['response_time'].fillna(value=data['response_time'].mean(), inplace=True)
```

```python
col_to_log = ['price', 'follower_count', 'products', 
              'rating_good', 'shop_location_encoded']
for col in col_to_log:
    data[col] = np.log(data[col])
```

```python
silhouette_coefficients = []
k_train = data[['follower_count', 'response_rate']]

kmeans_kwargs= {
    "init":"random",
    "n_init":10,
    "max_iter":300,
    "random_state":42
}

for k in tqdm(range(2, 11)):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(k_train)
    score = silhouette_score(k_train, kmeans.labels_)
    silhouette_coefficients.append(score)
    
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=2,
    n_init=10,
    max_iter=300,
    random_state=42
)

kmeans.fit(k_train)
train_cluster = kmeans.predict(k_train)
data['cluster'] = train_cluster
sns.scatterplot(x=data['response_rate'], y=data['follower_count'], hue=data['cluster'])
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, dpi=150, sharex=True, sharey=True)
for index, i in enumerate(data['cluster'].unique()):
    cluster = data[data['cluster'] == i].copy()
    cluster.drop(columns=['cluster'], inplace=True)
    train_data, test_data = train_test_split(cluster, test_size=0.1, random_state=42)
    if i == 0:
        model = RandomForestRegressor(n_estimators=100, max_depth=10, min_samples_split=3, min_samples_leaf=2)
    elif i == 1:
        model = RandomForestRegressor(n_estimators=50, max_depth=6, min_samples_split=2, min_samples_leaf=1)
    model.fit(train_data.drop(columns=['historical_sold']), train_data['historical_sold'])
    preds = model.predict(test_data.drop(columns=['historical_sold']))
    test_data['preds'] = preds
    
    MAE = mean_absolute_error(test_data['historical_sold'], test_data['preds'])
    MSE = mean_squared_error(test_data['historical_sold'], test_data['preds'])
    RMSE = MSE**0.5
    print('*'*20)
    print(f'Cluster: {i}')
    print(f'Num sample: {len(cluster)}')
    print(f"Mean target: {test_data['historical_sold'].mean():.2f}")
    print(f'MAE: {MAE:.2f}, MSE: {MSE:.2f}, RMSE: {RMSE:.2f}')
    
    sns.lineplot(data=test_data['historical_sold'], label='actual', ax=axes[index])
    sns.lineplot(data=test_data['preds'], label='pred', ax=axes[index])
# plt.show()
```
